```python
import matplotlib.pyplot as plt

sub_df1_grouped = df_batch1[["Prompt", "Correctness", "Consistency", "Completness", "Efficiency"]].groupby("Prompt").mean()
sub_df2_grouped = df_batch2[["Prompt", "Correctness", "Consistency", "Completness", "Efficiency"]].groupby("Prompt").mean()

# Define median properties for the box plots
medianprops = dict(linestyle='-', linewidth=3, color='firebrick')
fig, ax = plt.subplots(figsize=(8, 6))

# Number of criteria groups
n_groups = len(sub_df1_grouped.columns)

# Positions for the box plots
positions_1 = [1 + i*1.2 for i in range(n_groups)]
positions_2 = [1.3 + i*1.2 for i in range(n_groups)]

# Create the box plots for the first dataset
boxplots1 = ax.boxplot(
    [sub_df1_grouped[col].dropna().values for col in sub_df1_grouped.columns],
    positions=positions_1, widths=0.3, patch_artist=True,
    medianprops=medianprops
)

# Create the box plots for the second dataset
boxplots2 = ax.boxplot(
    [sub_df2_grouped[col].dropna().values for col in sub_df2_grouped.columns],
    positions=positions_2, widths=0.3, patch_artist=True,
    medianprops=medianprops
)

# Set colors for the box plots
for patch in boxplots1['boxes']:
    patch.set_facecolor('lightblue')
for patch in boxplots2['boxes']:
    patch.set_facecolor('lightgreen')

# Set x-axis labels at the center of each group of box plots
ax.set_xticks([1.15 + i * 1.2 for i in range(n_groups)])
ax.set_xticklabels(sub_df1_grouped.columns, fontsize=13)

# Set title and labels
ax.set_xlabel('Criteria Group', fontsize=13)
ax.set_ylabel('Score', fontsize=13)
ax.set_ylim([1.5, 5])

# Customize the grid
ax.yaxis.grid(True)  # Show horizontal grid lines
ax.xaxis.grid(False) # Hide vertical grid lines

# Set the legend
plt.legend([boxplots1["boxes"][0], boxplots2["boxes"][0]], ['Batch 1', 'Batch 2'], loc='upper right', fontsize = 13)

# Rotate x-axis labels if they overlap
# plt.xticks(rotation=90)
# for label in ax.get_xmajorticklabels():
#     label.set_rotation(45)
#     label.set_horizontalalignment("right")

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

# Adjust layout to prevent clipping
plt.tight_layout()

# Show the plot
plt.show()
```
